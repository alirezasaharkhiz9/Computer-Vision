# Speed Estimation With YOLO v11

```python
import cv2
from ultralytics import solutions

cap = cv2.VideoCapture("highway_mini.mp4")
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
video_writer = cv2.VideoWriter("speed_estimation1.avi", cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))
w, h
```

```python
speed_obj = solutions.SpeedEstimator(
    region=[(0, 450), (1900, 450)],
    model="yolo11s.pt",
    show=True,
)

while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        break
    im0 = speed_obj.estimate_speed(im0)
    video_writer.write(im0)

cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
