```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

## one shot learning with ArcFace model

```python
reference_img_path = "/content/Cristiano_Ronaldo,_2023.jpg"
test_img_path = "/content/cristiano_ronaldo_with_ucl_trophy_photo.jpeg"

def show_img(img_path, title):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

show_img(reference_img_path, "Reference Image")
show_img(test_img_path, "Test Image")
```

```python
result = DeepFace.verify(
    img1_path=reference_img_path,
    img2_path=test_img_path,
    model_name="ArcFace",
    detector_backend="retinaface",
    distance_metric="cosine",
    enforce_detection=True
)

print(f"Result: {result['verified']}")
print(f"Cosine Distance: {result['distance']:.4f}")
print(f"Threshold: {result['threshold']:.4f}")
```

## Face Analysis

```python
import pprint

img_path = "/content/Cristiano_Ronaldo,_2023.jpg"

obj = DeepFace.analyze(
    img_path=img_path,
    actions=['age', 'gender', 'race', 'emotion'],
    detector_backend="mtcnn",
    # enforce_detection=True
)

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

pp = pprint.PrettyPrinter(indent=4)
pp.pprint(obj)
```
