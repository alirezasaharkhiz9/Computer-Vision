# Heat Maps With Yolo v11

```python
import cv2

from ultralytics import solutions
```

```python
cap = cv2.VideoCapture("C:/Users/Alireza/Downloads/Documents/Undergraduate Project/speed/highway_mini.mp4")
assert cap.isOpened(), "Error reading video file"
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

video_writer = cv2.VideoWriter("heatmap_output.avi", cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))
```

```python
heatmap = solutions.Heatmap(colormap=cv2.COLORMAP_PARULA, show=True, model="yolo11s.pt")

while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        break
    im0 = heatmap.generate_heatmap(im0)
    video_writer.write(im0)
    

cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
